```python
#We will try to fill the null values by seeing the relationships between the expenses and other variables.
variables = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
pd.plotting.scatter_matrix(df[variables], figsize=(10, 10)) # Plotting the scatter matrix of the variables
```

```python
#We will calculate the correlation coeffecients to check for any patterns.
variables = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
corr_matrix = df[variables].corr()
print(corr_matrix)

#There doesn't seem to be a strong correlation between the variables. We will try to fill the null values with the mean of the column.
```

```python
#Filling the null values with the mean of the column
df['RoomService'].fillna(df['RoomService'].mean(), inplace=True)
df['FoodCourt'].fillna(df['FoodCourt'].mean(), inplace=True)
df['ShoppingMall'].fillna(df['ShoppingMall'].mean(), inplace=True)
df['Spa'].fillna(df['Spa'].mean(), inplace=True)
df['VRDeck'].fillna(df['VRDeck'].mean(), inplace=True)
```

```python
#Dropping duplicate values
df.drop_duplicates(inplace=True)
```

```python
#Checking for null values again to see if they have been filled
df.isnull().sum()
```

```python
#we will calcaulte the frequency of the non-numerical variables
planet_freq = df['HomePlanet'].value_counts()
cryo_freq = df['CryoSleep'].value_counts()
Cabin_freq = df['Cabin'].value_counts()
destination_freq = df['Destination'].value_counts()
vip_freq = df['VIP'].value_counts()

print("Planet Frequence", planet_freq)
print("CryoSleep Frequence", cryo_freq)
print("Destination Frequence", destination_freq)
print("VIP Frequence", vip_freq)
```

```python
#Drop the cabin column
df.drop('Cabin', axis=1, inplace=True)

#We decided to drop all the missing values, as they seem to be insignifcant in numbers.
df.dropna(inplace=True)
```

```python
#dropping the unecessary columns
df.drop(['PassengerId', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck'], axis=1, inplace=True)
```
